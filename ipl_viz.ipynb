```python
import pandas as pd
import seaborn as sns
df = pd.read_csv(r"D:\Desktop\ipl_cleaned.csv")
print(df.head())
df["match_date"] = pd.to_datetime(df["match_date"])
print(df.info())
```

```python
# =========================================
# IPL ANALYSIS - WINS BY TEAM
# =========================================

import matplotlib.pyplot as plt

# Count total wins by each team
team_wins = df["winner"].value_counts()

# Apply plotting style
plt.style.use("ggplot")

# Create figure
plt.figure(figsize=(12,6))

# Horizontal bar chart
plt.barh(
    team_wins.index,
    team_wins.values
)

# Chart title and labels
plt.title("Total Wins by Team", fontsize=16)

plt.xlabel("Number of Wins", fontsize=12)

plt.ylabel("Teams", fontsize=12)

# Display chart
plt.show()
```

```python
# =========================================
# IPL ANALYSIS - TOSS DECISION DISTRIBUTION
# =========================================

import matplotlib.pyplot as plt

# Count toss decisions
toss_decisions = df["toss_decision"].value_counts()

# Create figure
plt.figure(figsize=(6,6))

# Pie chart
plt.pie(
    toss_decisions.values,
    labels=toss_decisions.index,
    explode=(0, 0.1),
    autopct="%1.1f%%"
)

# Chart title
plt.title("Toss Decision Distribution", fontsize=16)

# Display chart
plt.show()
```

```python
# =========================================
# IPL ANALYSIS - TOP PLAYER OF THE MATCH WINNERS
# =========================================

import matplotlib.pyplot as plt
import seaborn as sns

# Count player of the match awards
top_award_winners = (
    df["player_of_match"]
    .value_counts()
    .head(10)
)

# Apply seaborn style
sns.set_style("darkgrid")

# Create figure
plt.figure(figsize=(12,6))

# Bar chart
sns.barplot(
    x=top_award_winners.index,
    y=top_award_winners.values
)

# Chart title and labels
plt.title(
    "Top 10 Player of the Match Winners",
    fontsize=16
)

plt.xlabel("Players", fontsize=12)

plt.ylabel("Number of Awards", fontsize=12)

# Rotate labels for readability
plt.xticks(rotation=45)

# Display chart
plt.show()
```

```python
# =========================================
# IPL ANALYSIS - SEASON WISE TEAM PERFORMANCE
# =========================================

from ipywidgets import interact
import matplotlib.pyplot as plt
import seaborn as sns

# Interactive season analysis function
def season_analysis(season):

    # Filter selected season data
    season_df = df[df["Season"] == season]

    # Total matches in selected season
    total_matches = season_df.shape[0]

    print(f"Total Matches in {season} : {total_matches}")

    # Count wins by team
    team_wins = season_df["winner"].value_counts()

    # Create figure
    plt.figure(figsize=(12,6))

    # Bar chart
    sns.barplot(
        x=team_wins.values,
        y=team_wins.index
    )

    # Chart title and labels
    plt.title(
        f"Team Wins in IPL Season {season}",
        fontsize=16
    )

    plt.xlabel("Number of Wins", fontsize=12)

    plt.ylabel("Teams", fontsize=12)

    # Display chart
    plt.show()

# Interactive dropdown
interact(
    season_analysis,
    season=sorted(df["Season"].unique())
)
```

```python
# =========================================
# IPL ANALYSIS - TOSS IMPACT ON MATCH RESULT
# =========================================

import matplotlib.pyplot as plt
import seaborn as sns

# Create new column to check
# whether toss winner also won the match
df["toss_match_winner"] = (
    df["toss_winner"] == df["winner"]
)

# Create figure
plt.figure(figsize=(7,5))

# Countplot
sns.countplot(
    x="toss_match_winner",
    data=df
)

# Chart title and labels
plt.title(
    "Did Toss Winner Also Win the Match?",
    fontsize=16
)

plt.xlabel(
    "Toss Winner Won Match",
    fontsize=12
)

plt.ylabel(
    "Number of Matches",
    fontsize=12
)

# Display chart
plt.show()
```

```python
# =========================================
# IPL ANALYSIS - MOST POPULAR VENUES
# =========================================

import matplotlib.pyplot as plt
import seaborn as sns

# Count matches played at each venue
venue_matches = (
    df["venue"]
    .value_counts()
    .head(5)
)

# Create figure
plt.figure(figsize=(12,6))

# Bar chart
sns.barplot(
    x=venue_matches.values,
    y=venue_matches.index
)

# Chart title and labels
plt.title(
    "Top 5 Most Popular IPL Venues",
    fontsize=16
)

plt.xlabel(
    "Number of Matches",
    fontsize=12
)

plt.ylabel(
    "Venue",
    fontsize=12
)

# Display chart
plt.show()
```

```python
# =========================================
# IPL ANALYSIS - TOTAL TOSS WINS BY TEAM
# =========================================

import matplotlib.pyplot as plt
import seaborn as sns

# Count total toss wins by team
toss_wins = df["toss_winner"].value_counts()

# Create figure
plt.figure(figsize=(12,6))

# Bar chart
sns.barplot(
    x=toss_wins.values,
    y=toss_wins.index
)

# Chart title and labels
plt.title(
    "Total Toss Wins by Team",
    fontsize=16
)

plt.xlabel(
    "Number of Toss Wins",
    fontsize=12
)

plt.ylabel(
    "Teams",
    fontsize=12
)

# Display chart
plt.show()
```

```python
# =========================================
# IPL ANALYSIS - SEASON WISE TOSS DECISIONS
# =========================================

from ipywidgets import interact
import matplotlib.pyplot as plt

# Interactive toss decision analysis
def toss_analysis(season):

    # Filter selected season
    season_df = df[df["Season"] == season]

    # Count toss decisions
    toss_decisions = (
        season_df["toss_decision"]
        .value_counts()
    )

    # Create figure
    plt.figure(figsize=(7,5))

    # Line plot
    plt.plot(
        toss_decisions.index,
        toss_decisions.values,
        marker="o"
    )

    # Chart title and labels
    plt.title(
        f"Toss Decisions in IPL Season {season}",
        fontsize=16
    )

    plt.xlabel(
        "Decision Type",
        fontsize=12
    )

    plt.ylabel(
        "Number of Matches",
        fontsize=12
    )

    # Display chart
    plt.show()

# Interactive dropdown
interact(
    toss_analysis,
    season=sorted(df["Season"].unique())
)
```

```python
# =========================================
# IPL ANALYSIS - CORRELATION HEATMAP
# =========================================

import matplotlib.pyplot as plt
import seaborn as sns

# Select numerical columns
numerical_columns = (
    df.select_dtypes(include="number")
)

# Display numerical column names
print("Numerical Columns:")
print(numerical_columns.columns)

# Create correlation matrix
correlation_matrix = numerical_columns.corr()

# Create figure
plt.figure(figsize=(10,6))

# Heatmap
sns.heatmap(
    correlation_matrix,
    annot=True,
    cmap="coolwarm"
)

# Chart title
plt.title(
    "Correlation Heatmap",
    fontsize=16
)

# Display chart
plt.show()

# Display correlation values
print("\nCorrelation Matrix:")
print(correlation_matrix)
```

```python
# =========================================
# IPL ANALYSIS - KPI DASHBOARD
# =========================================

import matplotlib.pyplot as plt

# Top winning team
top_team = df["winner"].value_counts().idxmax()

top_team_wins = (
    df["winner"]
    .value_counts()
    .max()
)

# Top player of the match winner
top_player = (
    df["player_of_match"]
    .value_counts()
    .idxmax()
)

top_player_awards = (
    df["player_of_match"]
    .value_counts()
    .max()
)

# Most hosting city
top_city = (
    df["city"]
    .value_counts()
    .idxmax()
)

top_city_matches = (
    df["city"]
    .value_counts()
    .max()
)

# Most preferred toss decision
top_toss_decision = (
    df["toss_decision"]
    .value_counts()
    .idxmax()
)
plt.rcParams["font.family"] = "Segoe UI Emoji"
# Create figure
fig, ax = plt.subplots(figsize=(14,5))

# Remove axes
ax.axis("off")

# KPI Dashboard Text
kpi_text = f"""
🏆 Top Winning Team
{top_team} ({top_team_wins} Wins)

⭐ Top Player of the Match
{top_player} ({top_player_awards} Awards)

🏟️ Most Hosting City
{top_city} ({top_city_matches} Matches)

🪙 Most Preferred Toss Decision
{top_toss_decision}
"""

# Display KPI text
ax.text(
    0.5,
    0.5,
    kpi_text,
    fontsize=14,
    ha="center",
    va="center",
    bbox=dict(
        facecolor="lightblue",
        edgecolor="black",
        boxstyle="round,pad=1"
    )
)

# Dashboard title
plt.title(
    "IPL KPI Dashboard",
    fontsize=18
)

# Display dashboard
plt.show()
```
